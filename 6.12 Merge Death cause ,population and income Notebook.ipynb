```python
# Extract relevant population data from df_api
population_columns = ['Country Code'] + [str(year) for year in df['Year'].unique()]
```

```python
df_population = df_api[population_columns].melt(id_vars=['Country Code'], var_name='Year', value_name='Population')
```

```python
df_population['Year'] = df_population['Year'].astype(int)
```

```python
# Rename 'Code' to 'Country Code' in df_wrangled
df.rename(columns={'Code': 'Country Code'}, inplace=True)
```

```python
# Merge population data with df_wrangled
df = pd.merge(df, df_population, on=['Country Code', 'Year'], how='left')
```

```python
# Extract relevant income group data from df_metadata
df_income = df_metadata[['Country Code', 'IncomeGroup']]
```

```python
# Merge income group data with df_wrangled
df = pd.merge(df, df_income, on='Country Code', how='left')
```
