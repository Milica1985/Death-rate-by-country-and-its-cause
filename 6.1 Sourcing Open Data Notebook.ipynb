```python
# Convert 'Year' column to integer type
df['Year'] = df['Year'].astype(int)
```

```python
# Defining Decade group 
df.loc[(df['Year'] >= 1990) & (df['Year'] <= 1999), 'Decade'] = '90s'
df.loc[(df['Year'] >= 2000) & (df['Year'] <= 2009), 'Decade'] = '00s'
df.loc[(df['Year'] >= 2010) & (df['Year'] <= 2019), 'Decade'] = '10s'
```

```python
#Checking output
df['Decade'].value_counts(dropna = False)
```

```python
# Define enviromental risk categories
environmental_factors = ['Outdoor air pollution', 'Unsafe water source', 'Household air pollution from solid fuels', 'Unsafe sanitation', 'No access to handwashing facility']
```

```python
df['Environmental Factors'] = df[environmental_factors].sum(axis=1)
```

```python
#Define life style risk category
lifestyle_factors = ['Alcohol use', 'Smoking', 'Low physical activity', 'Diet low in fruits', 'Diet low in Vegetables', 'Diet low in whole grains', 'Diet low in nuts and seeds','Diet high in sodium ']
```

```python

df['Lifestyle Factors'] = df[lifestyle_factors].sum(axis=1)
```

```python
#Define Health risk category
health_conditions = ['High systolic blood pressure', 'High fasting plasma glucose', 'High body mass index', 'Vitamin A deficiency', 'Iron deficiency']
```

```python
df['Health Conditions'] = df[health_conditions].sum(axis=1)
```

```python
#Define Maternal child risk category
maternal_child_health = ['Low birth weight','Child stunted growt','Child wasting', 'Discontinued breastfeeding', 'Non-exclusive breastfeeding']
```

```python
df['Maternal and Child Health'] = df[maternal_child_health].sum(axis=1)
```
