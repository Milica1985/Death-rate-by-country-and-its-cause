### Exploring relationships

```python
#Create a correlation matrix using pandas
#Select only the numerical columns from the DataFrame to avoid errors
numerical_columns = df.select_dtypes(include=['int', 'float'])

#The index ('rank') is one of the numerical columns, so we will drop it

if 'rank' in numerical_columns.columns:
    numerical_columns = numerical_columns.drop(columns=['rank'])

#Now, we create the correlation matrix with the appropriate data
numerical_columns.corr()
```

```python
#Now, we use the above data to create a correlation heatmap using matplotlib

plt.matshow(numerical_columns.corr())
plt.show()
```

```python
#Next, we add a legend, column labels, change the size of the heatmap and add a title
f=plt.figure(figsize=(9,9)) #This establishes the size of the heatmap
plt.matshow(numerical_columns.corr(), fignum=f.number) #Defines the type of plot
plt.xticks(range(numerical_columns.shape[1]), numerical_columns.columns, fontsize=12, rotation=90)
plt.yticks(range(numerical_columns.shape[1]), numerical_columns.columns, fontsize=12) #No need for rotation in y-axis
cb=plt.colorbar() #add a color legend (optimizes itseld based on colors used in heatmap)
cb.ax.tick_params(labelsize=12) #add font size
plt.title('Correlation Matrix', fontsize=15) #Creates titles
```

```python
#Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(numerical_columns.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

### The lighter the color, the stronger the (positive) relationship, and vice versa for the darker colors.

### Pair plots

```python
# Recreate the scatter plot for 'Outdoor air pollution' vs. 'High systolic blood pressure'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Outdoor air pollution', y='High systolic blood pressure')
sns.regplot(data=df, x='Outdoor air pollution', y='High systolic blood pressure', scatter=False, color='lightblue')
plt.title('Outdoor Air Pollution vs. High Systolic Blood Pressure')
plt.xlabel('Outdoor Air Pollution')
plt.ylabel('High Systolic Blood Pressure')
plt.show()
```

```python
 #Recreate the scatter plot for 'Diet high in sodium' vs. 'High systolic blood pressure'
# Scatter plot with trend line
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Diet high in sodium', y='High systolic blood pressure')
sns.regplot(data=df, x='Diet high in sodium', y='High systolic blood pressure', scatter=False, color='lightblue')
plt.title('Diet High in Sodium vs. High Systolic Blood Pressure')
plt.xlabel('Diet High in Sodium')
plt.ylabel('High Systolic Blood Pressure')
plt.show()
```

#### The scatter plot and trend line show the relationship between "Diet high in sodium" and "High systolic blood pressure, helping us understand if higher sodium intake is linked to higher blood pressure. This is valuable for dietary recommendations.



```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Alcohol use', y='High systolic blood pressure')
sns.regplot(data=df, x='Alcohol use', y='High systolic blood pressure', scatter=False, color='lightblue')
plt.title('Alcohol Use vs. High Systolic Blood Pressure')
plt.xlabel('Alcohol Use')
plt.ylabel('High Systolic Blood Pressure')
plt.show()
```

### Categorical Plot

```python
# Plotting the box plot for Country and Unsafe water source
plt.figure(figsize=(20, 10))
sns.boxplot(data=df, x='Country', y='Unsafe water source')
plt.title('Unsafe Water Source by Country')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Unsafe Water Source')
```

```python
# Create bins for Outdoor Air Pollution
df['Outdoor Air Pollution Category'] = pd.cut(df['Outdoor air pollution'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])

# Plotting the box plot for Outdoor Air Pollution categorized by Continents
plt.figure(figsize=(20, 10))
sns.boxplot(data=df, x='Outdoor Air Pollution Category', y='Outdoor air pollution', hue='Continents')
plt.title('Outdoor Air Pollution by Continent')
plt.xlabel('Outdoor Air Pollution Category')
plt.ylabel('Outdoor Air Pollution')
plt.legend(title='Continent')
plt.show()
```

#### Asia and Africa typically have higher pollution levels (Medium to Very High categories), indicating more significant pollution issues. Oceania and Europe usually have better air quality, falling within the Very Low and Low categories. â€¢Outliers (dots outside the whiskers) indicate regions with unusually high or low pollution levels compared to the overall trend.
